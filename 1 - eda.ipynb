### Look at where the capital letters are used

Most all-caps are healthcare acronyms like A&E, GP, COVID, ADHD, etc, which retain meaning in lower-case

Relatively few 'shouting caps' observed. In cases present, sometimes accompnied with punctuation.

Operate on assumption that lowercasing (to help limit vocab) is preferred to retaining case

```python
words_w_caps = list(filter(
    lambda w: w.strip().lower() != w.strip(),
    ' '.join(df_raw['FFT answer'].astype(str)).split()
))
sorted(words_w_caps)

words_w_caps
```

# 6. Appendix

### Can I balance the dataset by pruning the thin positive comments?
 - Thin comments don't make up much of the data. And in combination with question type (required), they are informative.

Suppose I prune the thin positive comments, what is the resulting distribution?

`Comment sentiment=1` drops only from 39% to 35%

```python
thin_pos_bool = df_raw['Comment sentiment'].eq(1) & df_raw['FFT answer'].str.len().lt(10)
df_raw['Comment sentiment'].loc[~thin_pos_bool].value_counts(normalize=True).sort_index().mul(100).round().rename('percent')
```

See how often thin comments are positive

 Most short comments are when sentiment=1. Overlaps with the mode for sentiment=5.

```python
df_tweak_sent.loc[[3937]].info()
```

```python
clip_lengths = 50 #zoom in on shorter end of range

df2 = (
    df_raw[['FFT answer', 'Comment sentiment']]
    .loc[lambda df_: df_['Comment sentiment'].notna()]
    .assign(answer_char_len=lambda df_: df_['FFT answer'].str.len())
    .loc[lambda df_: df_.answer_char_len.lt(clip_lengths)]

    .pipe(lambda df_: display(df_.sample(10)) or df_)
)

sns.displot(
    df2, row='Comment sentiment', x='answer_char_len',
    kind='hist', bins=clip_lengths, edgecolor='none',
    height=2, aspect=2,
)
```
